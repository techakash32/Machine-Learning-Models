# 📉 PCA — Principal Component Analysis — Quick Notes
---
> **Type:** Unsupervised | **Task:** Dimensionality Reduction | **Dataset:** `mall`

## 📌 Key Points
- Reduces features to fewer **principal components** (PCs)
- PCs capture **maximum variance** in data
- Used for: visualization, noise removal, speed up ML models
- Features must be **scaled** before PCA
- PCs are **uncorrelated** with each other
- `explained_variance_ratio_` → how much variance each PC captures
- Choose components where cumulative variance ≥ 95%

## 📊 Dataset: `mall`
| Column | Role |
|---|---|
| `Age`, `Annual Income`, `Spending Score` | All 3 features → reduce to 2 PCs |

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# 1. Load & Prepare
df = pd.read_csv("/content/mall - mall.csv")
df = df.drop(columns=['CustomerID', 'Genre'])
x = df.values   # use all 3 features: Age, Annual Income, Spending Score

# 2. Scale
sc = StandardScaler()
x_scaled = sc.fit_transform(x)

# 3. PCA — check explained variance first
pca_full = PCA()
pca_full.fit(x_scaled)
print("Explained Variance Ratio:", pca_full.explained_variance_ratio_.round(3))
print("Cumulative:", pca_full.explained_variance_ratio_.cumsum().round(3))

plt.figure(figsize=(7,4))
plt.bar(range(1, len(pca_full.explained_variance_ratio_)+1),
        pca_full.explained_variance_ratio_, alpha=0.7, label='Individual')
plt.step(range(1, len(pca_full.explained_variance_ratio_)+1),
         pca_full.explained_variance_ratio_.cumsum(), where='mid', label='Cumulative', color='red')
plt.title('Explained Variance by Component')
plt.xlabel('Principal Component')
plt.ylabel('Variance Ratio')
plt.legend()
plt.show()

# 4. Reduce to 2 PCs
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x_scaled)
print(f"\nShape before PCA: {x_scaled.shape}")
print(f"Shape after PCA : {x_pca.shape}")

# 5. Cluster in PCA space & Visualize
km = KMeans(n_clusters=4, random_state=42)
labels = km.fit_predict(x_pca)

colors = ['blue','green','red','cyan']
plt.figure(figsize=(8,6))
for i in range(4):
    plt.scatter(x_pca[labels==i,0], x_pca[labels==i,1], s=80, c=colors[i], label=f'Cluster {i+1}')
plt.title('PCA — 2D Projection with K-Means Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```

## 🔧 Key Attributes
| Attribute | Meaning |
|---|---|
| `explained_variance_ratio_` | Variance captured by each PC |
| `components_` | Direction of each PC (loadings) |
| `n_components_` | Number of PCs kept |

## 🥇 Remember
- **Always scale** before PCA
- Choose components where cumulative variance ≥ **95%**
- PCA is a preprocessing step — combine with clustering or classification
- `pca.inverse_transform()` → reconstruct original features
