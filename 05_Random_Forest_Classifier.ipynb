# 🌲🌲 Random Forest Classifier — Quick Notes
---
> **Type:** Supervised | **Task:** Classification | **Dataset:** `Social_Network_Ads`

## 📌 Key Points
- **Ensemble** of many Decision Trees
- Each tree trained on **random subset** of data (Bagging)
- Final prediction = **majority vote** of all trees
- Much less overfitting than single Decision Tree
- No feature scaling needed

## 📊 Dataset: `Social_Network_Ads`
| Column | Role |
|---|---|
| `Gender`, `Age`, `EstimatedSalary` | Features (X) |
| `Purchased` | Target (y) |

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

df = pd.read_csv("/content/Social_Network_Ads - Social_Network_Ads.csv")
lb = LabelEncoder()
df['Gender'] = lb.fit_transform(df['Gender'])

x = df.drop(columns=['Purchased'])
y = df['Purchased']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

rf = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)
rf.fit(x_train, y_train)
y_pred = rf.predict(x_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("Top Features:", sorted(zip(x.columns, rf.feature_importances_), key=lambda x: -x[1]))
```

## 🔧 Key Parameters
| Param | Meaning |
|---|---|
| `n_estimators` | Number of trees |
| `max_depth` | Max depth of each tree |
| `max_features` | Features per split |
| `bootstrap` | Sampling with replacement |

## 🥇 Remember
- More trees → more stable (diminishing returns after ~100)
- Random Forest > Single Decision Tree almost always
- `oob_score=True` → free validation score
