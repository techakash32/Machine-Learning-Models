# 👥 KNN Classifier — Quick Notes
---
> **Type:** Supervised | **Task:** Classification | **Dataset:** `Social_Network_Ads`

## 📌 Key Points
- **No training phase** — lazy learner
- Prediction = majority class among K nearest points
- Distance metric: **Euclidean** (default)
- **Must scale** features
- Large K → smoother boundary
- Small K → complex boundary (overfit)
- Slow on large datasets

## 📊 Dataset: `Social_Network_Ads`
| Column | Role |
|---|---|
| `Gender`, `Age`, `EstimatedSalary` | Features (X) |
| `Purchased` | Target (y) |

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

df = pd.read_csv("/content/Social_Network_Ads - Social_Network_Ads.csv")
lb = LabelEncoder()
df['Gender'] = lb.fit_transform(df['Gender'])

x = df.drop(columns=['Purchased'])
y = df['Purchased']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test  = sc.transform(x_test)

# Find best K
best_k, best_acc = 1, 0
for k in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    acc = accuracy_score(y_test, knn.predict(x_test))
    if acc > best_acc:
        best_acc, best_k = acc, k
print(f"Best K: {best_k}, Accuracy: {best_acc:.4f}")

knn = KNeighborsClassifier(n_neighbors=best_k)
knn.fit(x_train, y_train)
print(classification_report(y_test, knn.predict(x_test)))
```

## 🔧 Key Parameters
| Param | Meaning |
|---|---|
| `n_neighbors` | K — nearest neighbors count |
| `metric` | `euclidean`, `manhattan`, `minkowski` |
| `weights` | `uniform` or `distance` |

## 🥇 Remember
- Best K = usually **odd** to avoid ties
- Rule of thumb: K = √n
- `weights='distance'` often improves accuracy
