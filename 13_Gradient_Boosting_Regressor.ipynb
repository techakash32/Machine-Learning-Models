# 🚀 Gradient Boosting Regressor — Quick Notes
---
> **Type:** Supervised | **Task:** Regression | **Dataset:** `insurance`

## 📌 Key Points
- Sequential tree ensemble minimizing regression loss (MSE)
- Each tree fixes residual errors of previous
- One of the best out-of-the-box regressors
- Needs tuning to avoid overfit
- No feature scaling needed

## 📊 Dataset: `insurance`
| Column | Role |
|---|---|
| `age`, `bmi`, `children`, `sex`, `smoker`, `region` | Features (X) |
| `charges` | Target (y) |

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score, mean_absolute_error

df = pd.read_csv("/content/insurance - insurance.csv")
lb = LabelEncoder()
for col in ['sex','smoker','region']:
    df[col] = lb.fit_transform(df[col])

x = df.drop(columns=['charges'])
y = df['charges']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

gbr = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
gbr.fit(x_train, y_train)
y_pred = gbr.predict(x_test)

print("R2 Score:", r2_score(y_test, y_pred))
print("MAE     :", mean_absolute_error(y_test, y_pred))
```

## 🥇 Remember
- Usually better R² than Random Forest on tabular data
- `n_estimators=500` + `learning_rate=0.01` → often best combo
- Same params as Classifier — only loss function changes
