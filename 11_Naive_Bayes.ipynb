# 🎲 Naive Bayes — Quick Notes
---
> **Type:** Supervised | **Task:** Classification | **Dataset:** `Social_Network_Ads`

## 📌 Key Points
- Based on **Bayes' Theorem** with probability
- "Naive" = assumes all features are **independent**
- Very **fast** to train
- Works well with **text** data (spam detection)
- Three types: GaussianNB, MultinomialNB, BernoulliNB

## 🔑 Types
| Type | Use When |
|---|---|
| **GaussianNB** | Continuous features |
| **MultinomialNB** | Count data (text) |
| **BernoulliNB** | Binary features (0/1) |

## 📊 Dataset: `Social_Network_Ads`
| Column | Role |
|---|---|
| `Gender`, `Age`, `EstimatedSalary` | Features (X) |
| `Purchased` | Target (y) |

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

df = pd.read_csv("/content/Social_Network_Ads - Social_Network_Ads.csv")
lb = LabelEncoder()
df['Gender'] = lb.fit_transform(df['Gender'])

x = df.drop(columns=['Purchased'])
y = df['Purchased']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test  = sc.transform(x_test)

gnb = GaussianNB()
gnb.fit(x_train, y_train)
y_pred = gnb.predict(x_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("Class probabilities (first 3):", gnb.predict_proba(x_test[:3]))
```

## 🥇 Remember
- `P(class|features) ∝ P(features|class) × P(class)`
- Fast training — good baseline model
- Works despite "naive" assumption
- Not good when features are highly correlated
