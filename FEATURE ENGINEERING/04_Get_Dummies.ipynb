# 🐼 pd.get_dummies — Quick Notes
---
> **Dataset:** `tips` (seaborn) | **Library:** `pandas`

## 📌 Key Points
- **Pandas built-in** one-hot encoding (no sklearn needed)
- Simpler syntax than `OneHotEncoder`
- `drop_first=True` → removes first dummy to avoid multicollinearity
- Works directly on DataFrame — no reshape needed
- Automatically detects object/category columns
- Column names = `originalcol_value` format
- ⚠️ Unlike sklearn, **not fitted** — cannot transform new data the same way

## 🔑 get_dummies vs OneHotEncoder
| Feature | pd.get_dummies | OneHotEncoder |
|---|---|---|
| Syntax | Simple ✅ | More verbose |
| Pipeline ready | ❌ | ✅ |
| Save & reuse | ❌ | ✅ (fit once, transform many) |
| New data consistency | ❌ | ✅ |

```python
import numpy as np
import pandas as pd
import seaborn as sns

# 1. Load
df = sns.load_dataset('tips')
print("Original shape:", df.shape)
print(df.dtypes)
print()
df.head(3)
```

```python
# 2. get_dummies — basic usage
df_dummies = pd.get_dummies(df, columns=['sex', 'smoker', 'day', 'time'], drop_first=True)
df_dummies = df_dummies.astype(int)

print("After get_dummies shape:", df_dummies.shape)
df_dummies.head()
```

```python
# 3. See exactly which columns were created
original_cols = df.columns.tolist()
new_cols = df_dummies.columns.tolist()
added = [c for c in new_cols if c not in original_cols]
print("New dummy columns added:", added)
```

```python
# 4. Full ML Pipeline with get_dummies
df = sns.load_dataset('tips')
df = pd.get_dummies(df, columns=['sex', 'smoker', 'day', 'time'], drop_first=True)
df = df.astype(int)

x = df.drop(columns=['total_bill'])
y = df['total_bill']

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

lr = LinearRegression()
lr.fit(x_train, y_train)
print("R2 Score:", r2_score(y_test, lr.predict(x_test)))
print("Feature columns used:", list(x.columns))
```

## 🔧 Key Parameters
| Param | Meaning |
|---|---|
| `columns` | Which columns to encode (list) |
| `drop_first` | Drop first dummy to avoid multicollinearity |
| `prefix` | Custom prefix for column names |
| `dtype` | Data type of output (default float64) |
| `dummy_na` | Create column for NaN values too |

## 🥇 Remember
- Auto-detects object columns if `columns` not specified
- Always `df.astype(int)` after — converts True/False to 1/0
- For **training pipelines**, prefer `OneHotEncoder` (reusable)
- For **quick EDA/analysis**, `get_dummies` is faster to write
- `drop_first=True` → removes first category (reference class)
