# 🔥 OneHotEncoder — Quick Notes
---
> **Dataset:** `tips` (seaborn) | **Library:** `sklearn.preprocessing`

## 📌 Key Points
- Creates **binary (0/1) columns** for each category
- Eliminates **false ordering** problem of LabelEncoder
- `drop='first'` → removes one column to avoid **dummy variable trap** (multicollinearity)
- `sparse_output=False` → returns dense array (easier to work with)
- Best for **nominal features** (no natural order: colors, cities, gender)
- Increases number of columns → can cause **curse of dimensionality**

## 🔑 Dummy Variable Trap
| Without drop | With drop=first |
|---|---|
| sex_Male, sex_Female | sex_Female only |
| 2 cols for 2 values | 1 col for 2 values ✅ |

## 📊 Dataset: `tips`
| Column | Categories | New Columns (drop=first) |
|---|---|---|
| `sex` | Male, Female | sex_Male |
| `smoker` | Yes, No | smoker_Yes |
| `day` | Sat/Sun/Thur/Fri | day_Sat, day_Sun, day_Thur |
| `time` | Dinner, Lunch | time_Lunch |

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder

# 1. Load
df = sns.load_dataset('tips')
print("Original shape:", df.shape)
print(df.head(3))
```

```python
# 2. OneHotEncoder on categorical columns
ohe = OneHotEncoder(drop='first', sparse_output=False, dtype=np.int32)

encoded = ohe.fit_transform(df[['sex', 'smoker', 'day', 'time']])

# Get column names
col_names = ohe.get_feature_names_out(['sex', 'smoker', 'day', 'time'])
print("New columns:", col_names)
print("Encoded shape:", encoded.shape)

df_ohe = pd.DataFrame(encoded, columns=col_names)
df_ohe.head()
```

```python
# 3. Combine with numeric columns
df_numeric = df[['total_bill', 'tip', 'size']].reset_index(drop=True)
df_final = pd.concat([df_numeric, df_ohe], axis=1)
print("Final shape:", df_final.shape)
df_final.head()
```

```python
# 4. Compare: with drop vs without drop
ohe_no_drop = OneHotEncoder(drop=None, sparse_output=False, dtype=np.int32)
ohe_drop    = OneHotEncoder(drop='first', sparse_output=False, dtype=np.int32)

cols_nodrop = ohe_no_drop.fit_transform(df[['sex']]).shape[1]
cols_drop   = ohe_drop.fit_transform(df[['sex']]).shape[1]

print(f"sex column — without drop: {cols_nodrop} cols | with drop='first': {cols_drop} col")
print("\nWithout drop:")
print(ohe_no_drop.get_feature_names_out(['sex']))
print("\nWith drop='first':")
print(ohe_drop.get_feature_names_out(['sex']))
```

## 🔧 Key Parameters
| Param | Options | Meaning |
|---|---|---|
| `drop` | `None`, `'first'`, `'if_binary'` | Remove reference column |
| `sparse_output` | `True`/`False` | Dense vs sparse matrix |
| `dtype` | `np.float64`, `np.int32` | Output data type |
| `handle_unknown` | `'error'`, `'ignore'` | Unseen categories at test time |

## 🥇 Remember
- Always use `drop='first'` to avoid dummy variable trap
- `get_feature_names_out()` → get column names after encoding
- `sparse_output=False` → needed for DataFrame conversion
- `handle_unknown='ignore'` → useful when test set has unseen categories
- For pandas → use `pd.get_dummies()` (simpler syntax)
