# 🔢 Ordinal Encoder — Quick Notes
---
> **Dataset:** `tips` (seaborn) | **Library:** `sklearn.preprocessing`

## 📌 Key Points
- Encodes **multiple columns at once** (unlike LabelEncoder)
- You **define the order** yourself via `categories` parameter
- Gives **meaningful rank** to categories
- Perfect for features with **natural order** (e.g. Low < Medium < High)
- If `categories` not given → alphabetical order (risky!)
- Returns a **numpy array** by default → wrap in DataFrame

## 🔑 LabelEncoder vs OrdinalEncoder
| Feature | LabelEncoder | OrdinalEncoder |
|---|---|---|
| Columns at once | 1 only | Multiple ✅ |
| Custom order | ❌ | ✅ |
| Used for | Target (y) | Features (X) |
| sklearn compatible | Fit on 1D | Fit on 2D |

## 📊 Dataset: `tips` (categorical columns only)
| Column | Custom Order Defined |
|---|---|
| `sex` | Male → 0, Female → 1 |
| `smoker` | No → 0, Yes → 1 |
| `day` | Sat → 0, Sun → 1, Thur → 2, Fri → 3 |
| `time` | Dinner → 0, Lunch → 1 |

````python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import OrdinalEncoder

# 1. Load
df = sns.load_dataset('tips')
df = df.drop(columns=['total_bill', 'tip', 'size'])
df_original = df.copy()
print("BEFORE encoding:")
print(df.head(3))
````

````python
# 2. OrdinalEncoder with custom category order
oe = OrdinalEncoder(categories=[
    ['Male', 'Female'],          # sex
    ['No', 'Yes'],               # smoker
    ['Sat', 'Sun', 'Thur', 'Fri'],   # day
    ['Dinner', 'Lunch']          # time
])

oe_array = oe.fit_transform(df)
df_encoded = pd.DataFrame(oe_array, columns=df_original.columns)

print("AFTER OrdinalEncoder:")
df_encoded.head(5)
````

````python
# 3. See the categories learned
for i, col in enumerate(df_original.columns):
    print(f"{col:10s}: {list(oe.categories_[i])} → {list(range(len(oe.categories_[i])))}")
````

````python
# 4. Inverse Transform
decoded = oe.inverse_transform(oe_array[:5])
print("Decoded back to original:")
pd.DataFrame(decoded, columns=df_original.columns)
````

## ⚠️ Important
```python
# MUST pass 2D array (DataFrame or 2D numpy array)
oe.fit_transform(df[['col1', 'col2']])  # ✅
oe.fit_transform(df['col1'])            # ❌ 1D won't work
```

## 🥇 Remember
- Define `categories` list **in the same order as columns**
- Without custom categories → alphabetical (may be wrong!)
- `oe.categories_` → check what it learned
- Wrap result in `pd.DataFrame(..., columns=original.columns)` to keep column names
- Use for ordinal features: education level, ratings, satisfaction scores
