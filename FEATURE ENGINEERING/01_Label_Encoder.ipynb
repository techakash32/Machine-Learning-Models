# 🏷️ Label Encoder — Quick Notes
---
> **Dataset:** `tips` (seaborn) | **Library:** `sklearn.preprocessing`

## 📌 Key Points
- Converts **categorical text → integer numbers** (0, 1, 2, 3...)
- Works on **one column at a time**
- Order is **alphabetical** by default (no meaningful rank)
- ⚠️ Creates **false ordering** — ML model may think 2 > 1 > 0
- Best for: **target column (y)** or **tree-based models** (they don't care about order)
- NOT recommended for linear models with no natural order

## 🔑 When to Use
| Situation | Use LabelEncoder? |
|---|---|
| Target column (y) | ✅ Yes |
| Ordinal features (Low/Mid/High) | ✅ Yes |
| Nominal features (colors, cities) | ❌ Use OHE instead |
| Tree-based models (RF, XGB) | ✅ OK |

## 📊 Dataset: `tips`
| Column | Values | After Encoding |
|---|---|---|
| `sex` | Male, Female | 0, 1 |
| `smoker` | Yes, No | 1, 0 |
| `day` | Sat, Sun, Thur, Fri | 2, 3, 4, 0 |
| `time` | Dinner, Lunch | 0, 1 |

````python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

# 1. Load dataset
df = sns.load_dataset('tips')
print("BEFORE encoding:")
print(df.head(3))
print("\nCategorical columns:", df.select_dtypes('object').columns.tolist())
````

````python
# 2. Apply LabelEncoder to each categorical column
lb = LabelEncoder()

df['sex']    = lb.fit_transform(df['sex'])
df['smoker'] = lb.fit_transform(df['smoker'])
df['day']    = lb.fit_transform(df['day'])
df['time']   = lb.fit_transform(df['time'])

print("AFTER encoding:")
df.head(5)
````

````python
# 3. See mapping for each column
df2 = sns.load_dataset('tips')

for col in ['sex', 'smoker', 'day', 'time']:
    lb = LabelEncoder()
    lb.fit(df2[col])
    mapping = dict(zip(lb.classes_, lb.transform(lb.classes_)))
    print(f"{col:10s}: {mapping}")
````

````python
# 4. Inverse Transform — decode back to original
lb = LabelEncoder()
df2 = sns.load_dataset('tips')
lb.fit(df2['sex'])
encoded = lb.transform(df2['sex'])
decoded = lb.inverse_transform(encoded)
print("Encoded :", encoded[:5])
print("Decoded :", decoded[:5])
````

## ⚠️ Common Mistake
```python
# WRONG — applying on whole dataframe at once
lb.fit_transform(df)  # ❌ doesn't work on multiple columns

# RIGHT — column by column
df['col'] = lb.fit_transform(df['col'])  # ✅
```

## 🥇 Remember
- `fit_transform()` = fit + transform in one step
- `inverse_transform()` → decode numbers back to labels
- `lb.classes_` → see all unique categories found
- For multiple columns → loop or use `OrdinalEncoder` instead
