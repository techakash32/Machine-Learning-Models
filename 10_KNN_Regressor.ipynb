# 👥 KNN Regressor — Quick Notes
---
> **Type:** Supervised | **Task:** Regression | **Dataset:** `insurance`

## 📌 Key Points
- Prediction = **average** of K nearest neighbors' values
- Same idea as KNN Classifier — only output changes
- **Must scale** features
- No assumptions about data distribution
- Memory-intensive — stores entire training set

## 📊 Dataset: `insurance`
| Column | Role |
|---|---|
| `age`, `bmi`, `children`, `sex`, `smoker`, `region` | Features (X) |
| `charges` | Target (y) |

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score, mean_absolute_error

df = pd.read_csv("/content/insurance - insurance.csv")
lb = LabelEncoder()
for col in ['sex','smoker','region']:
    df[col] = lb.fit_transform(df[col])

x = df.drop(columns=['charges'])
y = df['charges']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test  = sc.transform(x_test)

# Find best K
best_k, best_r2 = 1, -999
for k in range(1, 21):
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(x_train, y_train)
    r2 = r2_score(y_test, knn.predict(x_test))
    if r2 > best_r2:
        best_r2, best_k = r2, k
print(f"Best K: {best_k}, R2: {best_r2:.4f}")

knn = KNeighborsRegressor(n_neighbors=best_k)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
print("R2 Score:", r2_score(y_test, y_pred))
print("MAE     :", mean_absolute_error(y_test, y_pred))
```

## 🥇 Remember
- Classifier = **vote** | Regressor = **average**
- Same K-finding loop works for both
- Scale is critical — large-range features dominate distance
